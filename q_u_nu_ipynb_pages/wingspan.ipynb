```python
import random
from collections import Counter

# Define dice faces
DICE_FACES = ["worm", "wheat", "fish", "berry", "rat"]

def roll_dice(num=5):
    return [random.choice(DICE_FACES) for _ in range(num)]

def all_same(dice):
    return len(set(dice)) == 1 and len(dice) > 0

def show_birdfeeder(birdfeeder):
    print("\nBirdfeeder state:")
    for face, count in Counter(birdfeeder).items():
        print(f"  {face}: {count}")
    print(f"  Total dice: {len(birdfeeder)}\n")

def take_die(birdfeeder, food_choice):
    if food_choice in birdfeeder:
        birdfeeder.remove(food_choice)
        print(f"Took a {food_choice} die.")
        return True
    else:
        print(f"No {food_choice} die available to take.")
        return False

def maybe_reroll_birdfeeder(birdfeeder):
    if all_same(birdfeeder):
        print(f"All remaining dice show '{birdfeeder[0]}'. Rerolling all remaining dice...\n")
        return roll_dice(len(birdfeeder))
    return birdfeeder

def example_turn():
    birdfeeder = roll_dice()
    show_birdfeeder(birdfeeder)

    while len(birdfeeder) > 0:
        choice = input("Which food would you like to take? (or 'stop' to end turn): ").capitalize()
        if choice.lower() == "stop":
            break

        valid = take_die(birdfeeder, choice)
        if valid:
            birdfeeder = maybe_reroll_birdfeeder(birdfeeder)
            show_birdfeeder(birdfeeder)

    print("Turn ended.")

if __name__ == "__main__":
    example_turn()
```
